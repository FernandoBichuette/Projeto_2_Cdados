```python
#importando as bibliotecas 
%matplotlib inline
import pandas as pd

import numpy as np
import seaborn as sns
from matplotlib.ticker import StrMethodFormatter
import statsmodels.formula.api as smf
from scipy import stats
from matplotlib.gridspec import GridSpec
import sklearn
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
```

```python
#Importando o dataset desemprego por escolaridade
desemprego_por_escolaridade = pd.read_csv("oecd.csv")
desemprego_por_escolaridade.drop(['INDICATOR','MEASURE','FREQUENCY','Flag Codes'], axis=1, inplace=True)
desemprego_por_escolaridade.head()
```

```python
#Importando o dataset desemprego total no ano
desemprego= pd.read_excel('desemprego_total.xls')
desemprego.drop(['Indicator Name','Indicator Code','Country Name'], axis=1, inplace=True)
desemprego.head()
```

```python
#Cruza as duas tabelas
dados=desemprego_por_escolaridade.merge(desemprego, left_on=['LOCATION','TIME'], right_on=['Country Code','YEARS'], how='right')
dados.drop(['YEARS','Country Code'], axis=1, inplace=True)
dados.head()
```

```python
#Renomeando o dataset para mais facil compreensão 
dados.rename(columns={'SUBJECT':'Escolaridade','Unemployment, total (% of total labor force)':'Desemprego'}, inplace=True)
dados["Escolaridade"]= dados["Escolaridade"].str.replace("BUPPSRY", "Prim_sem_Secun", case = False)
dados["Escolaridade"]=dados["Escolaridade"].str.replace("UPPSRY_NTRY", "Com_Secun_sem_superior", case = False) 
dados["Escolaridade"]=dados["Escolaridade"].str.replace("TRY", "Com_ens_superior", case = False) 
dados["LOCATION"]=dados["LOCATION"].str.replace("DEU", "Alem", case = False)
dados["LOCATION"]=dados["LOCATION"].str.replace("USA", "Eua", case = False)
dados["LOCATION"]=dados["LOCATION"].str.replace("KOR", "Coreia", case = False)
dados.head()
```

```python
#Fasendo um violinplot 
#1° analise exploratória
plt.figure(1, figsize=(10,5))

sns.violinplot(
    data=dados,
    x="LOCATION",
    y='Value',
    hue='Escolaridade',
    kind='box',
    height=50, 
    aspect=50) 
plt.title('Escolaridade x Porcentagem de desempregados')
plt.grid(True)
plt.show()
```

**1° análise exploratória:** Após os processos de limpeza dos dois datasets e a sua junção, foi feito um violin plot com os dados de desemprego por escolaridade ao longo de 21 anos nos três paises escolhidos. O grafico mostra algumas coisas bem interessantes. 

Se olharmos para Coreia do Sul sozinha podemos ver que existe pouca diferença de variabilidade de desemprego entre as escolaridades. Também e possivel ver que existe uma quantidade grade de outliers nos EUA entre pessoas desempregadas que só tem primario. Pode ser visto tambem que na Alemanha passou por um momento de grande taxa de desemprego que refletiu em todas as escolaridas principalmente no nas pessoas que só possuem primario.

```python
# Preparando o database para ser feita a regressão
dados = pd.get_dummies(dados, columns=['LOCATION','Escolaridade'])
dados.head()
```

```python
#Dividindo o database 
y=dados["Value"] # O que quero prever (desemprego)
x=dados.drop('Value',axis=1)

y.head()
```

```python
x.head()
```

```python
#Separando x e y em teste e train
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=152)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
# Fazendo regressão linear
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score


linreg = LinearRegression()
linreg.fit(x_train,y_train)

y_pred_linerReg=linreg.predict(x_test)

print('MSE:',np.mean((y_test-y_pred_linerReg)**2))
print('MAPE',mean_absolute_error(y_test,y_pred_linerReg))
print('R²:',r2_score(y_test,y_pred_linerReg))
```

Após feita a **regressão linear** foram calculados o erro quadrático médio,erro absoluto médio e o R². 

O **erro quadrático médio** obtido foi de 4,838, ele não é um numero ideal, não esta proximo de 0, porém não é alto suficiente para descartar totalmente.

O **erro absoluto médio** é de 1,698, assim como o MSE o valor e um pouco alto do ideal porém não o suficiente para descartar

O **R²** é de 0,748. Esse valor obtido embora seja bem abaixo de 1, que seria o perfeito, ele é acima de 0,7 o que é o limite que o grupo acha adequado. 

```python
import numpy as np
import statsmodels.api as sm

model = sm.OLS(y,x)
results = model.fit()
results.summary()
```

```python
p_values = results.summary2().tables[1]['P>|t|']
print('Valor P:',p_values)
```

Podemos ver que os valores P são abaixo de 5% ou seja nos não rejeitamos H0 

```python
# Fitting Random Forest Regression to the dataset 
# import the regressor 
from sklearn.ensemble import RandomForestRegressor 
  
 # create regressor object 
model = RandomForestRegressor(n_estimators = 500, random_state=152) 
  
# fit the regressor with x and y data 
model.fit(x_train,y_train) 

y_pred_RFR = model.predict(x_test)

print('MSE:',mean_squared_error(y_test,y_pred_RFR))
print('MAPE',(mean_absolute_error(y_test,y_pred_RFR)))
print('R²:',r2_score(y_test,y_pred_RFR)) #R²
```

Após feita a **Random Forest Regression** foram calculados o erro quadrático médio,erro absoluto médio e o R². 

O **erro quadrático médio** obtido foi de 0.992, esse valor ficou bem mais proximo de de 0, (o ideal), comparado com o modelo de regressão linear.


O **erro absoluto médio** é de 0.611, esse valor mais baixo comparado com o modelo de regressão linear reforça a ideia que a Random Forest Regression prediz melhor do que o Linear Regression


O **R²** é de 0,948. Esse valor obtido bem proximo de 1 reforça o que foi escrito acima.

```python
from sklearn.ensemble import GradientBoostingRegressor

reg = GradientBoostingRegressor(random_state=152)
reg.fit(x_train,y_train)
Grad_pred=reg.predict(x_test)

print('MSE:',mean_squared_error(y_test,Grad_pred))
print('MAPE',mean_absolute_error(y_test,Grad_pred))
print('R²:',r2_score(y_test,Grad_pred)) #R²
```

Após feita a **Gradient Boosting Regressor** foram calculados o erro quadrático médio,erro absoluto médio e o R². 

O **erro quadrático médio** obtido foi de 0.388, esse valor ficou bem mais proximo de de 0, (o ideal), comparado com o modelo Random Forest Regression que ate então tinha sido o mais proximo do que o de regressão linear.


O **erro absoluto médio** é de 0.456, esse valor mais baixo comparado com o modelo de Random Forest Regression reforçando a ideia que a Gradient Boosting Regressor prediz melhor do que o Random Forest Regression e Linear Regression


O **R²** é de 0.9798. Esse valor obtido bem proximo de 1 reforça o que foi escrito acima.

```python
#Residos

sns.residplot(y_pred_linerReg,y_test,color='r',label='Linear Regression')
sns.residplot(y_pred_RFR,y_test,color='green',label='Random Forest Regression')
sns.residplot(Grad_pred,y_test,color='orange',label='Gradient Boosting Regressor')

plt.title('Residos dos três modelos')
plt.grid(True)
plt.legend()
plt.show()
```

Fazendo uma análise de resíduos podemos ver que dos três modelos o Gradient Boosting Regressor e o que melhor aproxima do y=0 ou seja, ele é o melhor modelo de regressão para o problema.
